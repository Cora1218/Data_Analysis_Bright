```python
import sqlite3
import pandas as pd
```

```python
# Conectar a la base de datos SQLite
conn = sqlite3.connect('D:/Proyecto Bright/bright.sqlite')

# Crear un cursor para ejecutar comandos SQL
cursor = conn.cursor()
```

```python
# Ejemplo de consulta SQL
cursor.execute('SELECT * FROM event')
data = cursor.fetchall()

# Obtener los nombres de las columnas
columns = [desc[0] for desc in cursor.description]

# Combinar los nombres de las columnas con los datos
result = [columns] + data
```

```python
# Crear un DataFrame a partir del resultado de la consulta
df = pd.DataFrame(result)
df
```

```python
# Guardar el DataFrame como un archivo CSV
df.to_csv('event.csv', header=False, index=False)
```
